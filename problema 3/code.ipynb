```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

IMG_NAME = 'img.jpg'
```

```python
def getPath(img):
    current_dir = os.getcwd()
    parent_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
    image_path = os.path.join(parent_dir, img)
    return image_path
```

```python
def plot_image_in_3d(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)    

    if image is None:
        raise ValueError(f"No se pudo cargar la imagen desde la ruta: {image_path}")
    
    height, width = image.shape

    X, Y = np.meshgrid(np.arange(width), np.arange(height))
    
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')    
    ax.plot_surface(X, Y, image, cmap='gray', rstride=1, cstride=1, linewidth=0, antialiased=False)
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Intensidad del Color Gris')
    
    plt.show()


```

```python
image_path = getPath(IMG_NAME)
print(image_path)
plot_image_in_3d(image_path)
```
