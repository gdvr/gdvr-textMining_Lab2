```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

IMG_NAME = 'img2.jpg'
```

```python
def getPath(img):
    current_dir = os.getcwd()
    parent_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
    image_path = os.path.join(parent_dir, img)
    return image_path
```

```python
def plot_images(original, filtered, title_filtered):
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.title('Imagen Original')
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.title(title_filtered)
    plt.imshow(cv2.cvtColor(filtered, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    
    plt.show()
```

```python
def apply_convolution(image, kernel):
    filtered_image = cv2.filter2D(image, -1, kernel)
    return filtered_image

def apply_filter(image_path, kernel):
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"No se pudo cargar la imagen desde la ruta: {image_path}")
    
    filtered_image = apply_convolution(image, kernel)
    plot_images(image, filtered_image, 'Imagen Filtrada')
```

```python
image_path = getPath(IMG_NAME)
print(image_path)


gaussian_kernel = cv2.getGaussianKernel(ksize=5, sigma=1)
gaussian_kernel = np.outer(gaussian_kernel, gaussian_kernel)
kernels= [
        {'gaussian': gaussian_kernel},
        {'sobel':  np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])},
        {'sharpen':   np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])},
        {'laplacian':   np.array([[0, 1, 0],
                             [1, -4, 1],
                             [0, 1, 0]])}    
]


for kernel in kernels:
    kernel_type = list(kernel.keys())
    print(kernel_type[0])
    apply_filter(image_path, kernel[kernel_type[0]])
```
