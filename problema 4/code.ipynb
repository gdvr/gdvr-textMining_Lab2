```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

IMG_NAME = 'img2.jpg'
ALPHA=1.2
BETA=30
GAMMA=2.0
C=1.0
```

```python
def getPath(img):
    current_dir = os.getcwd()
    parent_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
    image_path = os.path.join(parent_dir, img)
    return image_path
```

```python
def transform_image(image_path, tipo):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        raise ValueError(f"No se pudo cargar la imagen desde la ruta: {image_path}")
    
    if tipo == 'Negative':
        return 255 - image
    elif tipo == 'Linear':
         return cv2.convertScaleAbs(image, alpha=ALPHA, beta=BETA)
    elif tipo == 'Logarithmic':
        image_log = C * np.log1p(image)
        return image_log
    elif tipo == 'Exponential':
        image_exp = np.power(image / 255.0, GAMMA)        
        image_exp = np.uint8(cv2.normalize(image_exp, None, 0, 255, cv2.NORM_MINMAX))
        return image_exp
    else:
        raise ValueError("Tipo de conversi√≥n no soportado")        

```

```python
image_path = getPath(IMG_NAME)
print(image_path)

imagen = 'img.jpg'
metodos = ['Negative','Linear','Logarithmic','Exponential']

plt.figure(figsize=(15, 10))

i = 0
for metodo in metodos:    
    plt.subplot(2, 3, i + 1)
    plt.title(metodo)
    plt.imshow(transform_image(image_path,metodo), cmap='gray')
    plt.axis('off')
    i = i + 1
plt.tight_layout()
plt.show()
    
```
